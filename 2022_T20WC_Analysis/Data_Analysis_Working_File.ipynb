# Building the Team of Tournament

### Step 1: Get Top Order Betters (1, 2, 3)

#### Criteria For Selection 

RUNS >= 200 & Average >= 40 & Strike Rate >= 140

```python
bat_df = bat_df[bat_df['No. Of Innings']>=4]
```

```python
bat_df.head()
```

```python
def is_top_order_batter(pos):
    return pos in (1, 2, 3)
top_order_batters = bat_df[bat_df['Bat_Pos'].apply(is_top_order_batter)]
```

```python
TOB = top_order_batters[(top_order_batters['Runs_Scored'] >= 200) & (top_order_batters['Avg'] >= 40) & (top_order_batters['SR'] >= 135)]
TOB
```

### Step 2: Get Middle Order Batter (4, 5)

#### Criteria For Selection 

RUNS >= 150 & Average >= 35 & Strike Rate >= 150

```python
middle_order_batters = bat_df[(bat_df['Bat_Pos'] == 4) | (bat_df['Bat_Pos'] == 5)]
middle_order_batters.head()
```

```python
MOB = middle_order_batters[(middle_order_batters['Runs_Scored'] >= 150) & (middle_order_batters['Avg'] >= 35) & (middle_order_batters['SR'] >= 150)]
MOB
```

### Step 3: Get All Rounders

#### Criteria For selection
Runs >= 80 and Avg >= 20 and Wickets >= 10

```python
all_rounders = pd.read_csv('Cleaned_DataSet/all_rounder_data_for_whole_tournament.csv')
```

```python
all_rounders.head()
```

```python
AR = all_rounders[(all_rounders['Runs_Scored'] >= 80) & (all_rounders['Avg'] >= 20) & (all_rounders['Wickets'] >= 10)]
AR
```

### Step 4: Get Bowlers

Criteria For selection is Wickets Taken Per Inning

```python
bow_df.head()
```

```python
bow_df= bow_df[bow_df['No. Of Innings']>= 4]
```

```python
bow_df['Wikt per ing'] = bow_df['Wickets'] / bow_df['No. Of Innings']
```

```python
Bowlers = bow_df.sort_values(by= ['Wikt per ing', 'Economy'], ascending = [False, True]).head(4)
Bowlers
```
