### Batting Dataset

```python
batting_df = pd.read_csv('Scrapped_Data/Scrapped_Batting_ScoreCard.csv')
```

```python
batting_df.head()
```

##### Formatting Match_id and making it index of df

```python
batting_df['Match_ID'] = batting_df['Match_ID'].apply(lambda x: int(x.strip().split()[-1]))
```

```python
batting_df.head()
```

##### Making a Col of Out or Not Out Based on Dismissal Col

```python
def get_out_or_notOut(dismissal):
    if dismissal.strip().lower() == 'not out':
        return 'Not Out'
    else:
        return 'Out'
```

```python
batting_df['Out/Not_Out'] = batting_df['Dismissal'].apply(get_out_or_notOut)
```

```python
batting_df.head()
```

```python
batting_df.shape
```

##### Removing Dismissal Col as it is not much relevent

```python
batting_df.drop('Dismissal', axis = 1, inplace = True)
```

```python
batting_df.head()
```

##### Minutes Spent is not a relevent Criteria for analysis in T20 CRICKET, so Removing it

```python
batting_df.drop('Minutes Spent', axis = 1, inplace = True)
```

```python
batting_df.head()
```

##### Renaming Some Cols

```python
batting_df.rename(columns = {'Batting Position': 'Bat_Pos', 'Batsman_Name': 'Bat_Name', 'Strike Rate':'SR', 'Runs':'Runs_Scored'}, inplace = True)
```

```python
batting_df.head()
```

```python
batting_df.info()
```

```python
batting_df.info()
```

##### Converting SR to float

```python
batting_df[batting_df['SR'] == '-']
```

```python
### Replacing these SR with 0
def parse_SR(strike_rate):
    try:
        val = float(strike_rate.strip())
        return val
    except ValueError:
        return 0

batting_df['SR'] = batting_df['SR'].apply(parse_SR)
```

```python
batting_df.info()
```

```python
def parse_name(name):
    name = name.split('(')[0]
    name = name.split('†')[0]
    name = name.split('\xa0')[0]
    return name
```

```python
batting_df['Bat_Name'] = batting_df['Bat_Name'].apply(parse_name)
```

##### Saving Cleaned Data

```python
batting_df.to_csv('Cleaned_DataSet/MatchWise_Batting_Data.csv')
```

## Data PreProcessing

#### Get the Batting and Bowling Data of individual for whole Tournament

# FOR BATTING DATA

```python
bat_df = pd.read_csv('Cleaned_DataSet/MatchWise_Batting_Data.csv')
```

```python
bat_df.groupby('Bat_Name').get_group('Virat Kohli')
```

## What's The Plan

We Need to make a Dataset for each of the player for whole tournament with following columns

(Bat_Name, Runs, 4s, 6s, SR, Team, Bat_Pos, Innings, Not Outs, Average)

#### STEP 1. GETTING NAMES, TEAM, BATTING_POSITION

```python
df1 = bat_df.groupby("Bat_Name").aggregate(min)[['Team', 'Bat_Pos']]
df1.head()
```

##### STEP 2. GETTING NO OF INNINGS

```python
df2 = bat_df.groupby("Bat_Name").describe()['Match_ID'][['count']]
df2.head()
```

##### STEP 3. GETTING RUNS, 4S, 6S, Balls

```python
df3 = bat_df.groupby("Bat_Name").aggregate(sum)[['Runs_Scored', '4s', '6s', 'Balls']]
df3.head()
```

##### STEP 4. GETTING NO OF NOT OUTS AND OUTS

```python
def get_not_out_data(df):
    res = []
    batsmans_Names = df['Bat_Name']

    for batter in batsmans_Names:
        data = dict(df.groupby('Bat_Name').get_group(batter)['Out/Not_Out'].value_counts())
        data['Bat_Name'] = batter
        res.append(data)
    return res
```

```python
Not_out_Data = get_not_out_data(bat_df)
```

```python
Not_out_Data[:5]
```

```python
df4 = pd.DataFrame(Not_out_Data).groupby('Bat_Name').aggregate(max)
df4.head()
```

```python
df4.fillna(0, inplace = True)
```

```python
df4 = df4.astype(int)
df4.head()
```

##### STEP 5: MERGE ALL THE DATAFRAMES TO GET THE REQUIRED DATASET

```python
tournament_batting_data = df1.join([df2, df3, df4], how = 'outer')
tournament_batting_data.head()
```

##### STEP 6: CLEAN THE DATA FOR USAGE

###### Rename the 'count' to 'No of Innings' and converting it into 'int'

```python
tournament_batting_data['count'] = tournament_batting_data['count'].astype(int)
tournament_batting_data.rename(columns = {'count':'No. Of Innings'}, inplace = True)
```

```python
tournament_batting_data.head(3)
```

###### Getting the strike rate

```python
tournament_batting_data['SR'] = np.round(tournament_batting_data['Runs_Scored']*100 / tournament_batting_data['Balls'], 2)
```

```python
tournament_batting_data.head()
```

###### Get the Average of All Batters

```python
tournament_batting_data['Avg'] = tournament_batting_data['Runs_Scored']/tournament_batting_data['Out']
```

```python
tournament_batting_data.head()
```

```python
tournament_batting_data.info()
```

```python
tournament_batting_data.dropna(inplace = True)
```

```python
tournament_batting_data.head()
```

```python
tournament_batting_data.loc['Virat Kohli']
```

#### STEP 7. SAVE THIS FILE

```python
tournament_batting_data.to_csv('Cleaned_DataSet/player_batting_stats_for_whole_tournament.csv')
```

# FOR BOWLING DATA

### What's The Plan
We Need to make a Dataset for each of the player for whole tournament with following columns

(Bowler_Name, Team, No. Of Innings, Overs, Maidens, Runs, Wickets, Dots)

```python
bow_df = pd.read_csv('Cleaned_DataSet/MatchWise_Bowling_Data.csv')
bow_df.head()
```

#### STEP 1. GETTING NAMES, TEAM

```python
dff1 = bow_df.groupby("Bowler_Name").aggregate(min)[['Team']]
dff1.head()
```

#### STEP 2. GETTING NO OF INNINGS

```python
dff2 = bow_df.groupby("Bowler_Name").describe()['Match_ID'][['count']]
dff2.head()
```

#### STEP 3. GETTING Overs, Maidens, Runs, Wickets, Economy, Dots

```python
dff3 = bow_df.groupby("Bowler_Name").aggregate(sum)[['Overs', 'Maidens', 'Runs_Conceded', 'Wickets', 'Economy', 'Dots']]
dff3.head()
```

#### STEP 4: MERGE ALL THE DATAFRAMES TO GET THE REQUIRED DATASET

```python
tournament_bowling_data = dff1.join([dff2, dff3], how = 'outer')
tournament_bowling_data.head()
```

#### STEP 5: CLEAN THE DATA FOR USAGE

#### Rename the 'count' to 'No of Innings' and converting it into 'int'

```python
tournament_bowling_data['count'] = tournament_bowling_data['count'].astype(int)
tournament_bowling_data.rename(columns = {'count':'No. Of Innings'}, inplace = True)
```

```python
tournament_bowling_data.head()
```

#### Getting the 'Actual Economy' (so divide by no of innings)

```python
tournament_bowling_data['Economy'] = np.round(tournament_bowling_data['Runs_Conceded'] / tournament_bowling_data['Overs'], 2)
```

```python
tournament_bowling_data.head()
```

```python
tournament_bowling_data.info()
```

#### STEP 6. SAVE THIS FILE

```python
tournament_bowling_data.to_csv('Cleaned_DataSet/player_bowling_stats_for_whole_tournament.csv')
```
